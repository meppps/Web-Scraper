## Loop through

```python
# Loop Through
recordCount = 1
companySites = []

for url in urls:
    print(f"Processing #{recordCount}")
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    
    # Get links
    attributes = soup.find_all("a", href=lambda value: value and value.startswith("/url?q="))
    aList = []
    
    for a in attributes:
        a = a['href'].split('/url?q=')[1]
        aList.append(a)

    # Clean them    
    newList = []
    for a in aList:
        if '.com' in a:
            a = a.split('.com')[0]
            a += '.com'
        elif '.net' in a:
            a = a.split('.net')[0]
            a += '.net'
        elif '.org' in a:
            a = a.split('.org')[0]
            a += '.org'
        elif '.io' in a:
            a = a.split('.io')[0]
            a += '.io'
        newList.append(a)
     
    # Filter
    cleanList = []
    for a in newList:
        if 'amazon' not in a:
            if 'youtube' not in a:
                if 'yelp' not in a:
                    if 'facebook' not in a:
                        if 'wikipedia'not in a:
                            if 'groupon' not in a:
                                cleanList.append(a)
                                if 'yellowpages' not in a:
                                    if 'angieslist' not in a:
                                        if 'bbb' not in a:
                                            cleanList.append(a)                                   
                            else:
                                continue
     
    # Select best guess
    companySite = cleanList[0]
    companySites.append(companySite)    
    
    print(f"Company Site: {companySite}")
    print(f"--------------")
    
    
    
    recordCount +=1 
```

```python
df['sites'] = companySites
df = df[['name','sites']]
df.head(15)
```
